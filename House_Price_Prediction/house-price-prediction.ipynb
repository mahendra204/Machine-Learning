```python
from copy import deepcopy
train_data_sub= deepcopy(train_data[['MSSubClass','LotFrontage','SalePrice']])
train_data_sub
```

```python
train_data_sub=train_data_sub.fillna(train_data_sub.mean())
train_data_sub
```

```python
Train_X=train_data_sub.to_numpy()[:,:-1]
Train_Y=train_data_sub.to_numpy()[:,-1]
Train_Y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(Train_X,Train_Y)
model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
60*7+65*1260 + 88327
```

```python
train_data_sub['linear_predictions']=model.predict(Train_X)
train_data_sub
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(train_data_sub['SalePrice'], train_data_sub['linear_predictions'])
```
