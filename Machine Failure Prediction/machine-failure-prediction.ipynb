```python
import numpy as np
import pandas as pd 

import os
for dirname, x, files in os.walk('/kaggle/input'):
    for i in files:
        print(os.path.join(dirname, i))
```

```python
train='/kaggle/input/playground-series-s3e17/train.csv'
test='/kaggle/input/playground-series-s3e17/test.csv'
sub='/kaggle/input/playground-series-s3e17/sample_submission.csv'
```

```python
traindata=pd.read_csv(train)
traindata
```

```python
testdata=pd.read_csv(test)
testdata
```

```python
subdata=pd.read_csv(sub)
subdata
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score
```

```python
traindata.columns
```

```python
subdata.columns
```

```python
features=['Type', 'Air temperature [K]',
       'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]',
       'Tool wear [min]', 'TWF', 'HDF', 'PWF', 'OSF',
       'RNF']
```

```python
xtrain=traindata[features]
```

```python
ytrain=traindata['Machine failure']
```

```python
ytrain.shape
```

```python
xtrain, xval, ytrain, yval=train_test_split(xtrain, ytrain, test_size=0.2, random_state=1)
```

```python
xtrain
```

```python
xtrain.isnull().sum()
```

```python
xtrain=pd.get_dummies(xtrain)
xval=pd.get_dummies(xval)
```

```python
print(xtrain.shape, xval.shape)
```

```python
model=LogisticRegression()
model1=DecisionTreeClassifier(max_depth=11, random_state=1)
model2=RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
models=[model, model1, model2]
for i in models:
    print(i)
```

```python
logmodel=model.fit(xtrain,ytrain)
dtrmodel=model1.fit(xtrain,ytrain)
rfmodel=model2.fit(xtrain,ytrain)
```

```python
logpred=model.predict(xval)
dtrpred=dtrmodel.predict(xval)
rfpred=rfmodel.predict(xval)
```

```python
preds=[logpred,dtrpred,rfpred]
for i in range(len(preds)):
    print(f'roc_auc_score {roc_auc_score(yval, preds[i])}')
```

****We can observe the value of roc auc score for random forest classifer is higher so that indicates accurate model than other two models. lets check.

```python
testd=testdata[features]
testd=pd.get_dummies(testd)
```

```python
logpred=model.predict(testd)
dtrpred=dtrmodel.predict(testd)
rfpred=rfmodel.predict(testd)
```
