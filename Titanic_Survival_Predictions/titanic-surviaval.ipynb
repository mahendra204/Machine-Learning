```python
train_data=pd.read_csv("/kaggle/input/titanic/train.csv")
```

```python
train_data.columns
```

```python
features = ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
train_data=train_data[features]
```

```python
train_data.dropna(inplace=True)
```

```python
train_data
```

```python
# train_data
# 3 ->[0,0,1]
# 2 -> [0,1,0]
# 1 -> [0,0,1]
# male -> 1 
# female -> 0
```

```python
from tensorflow.keras.utils import to_categorical
Pclass = train_data['Pclass'].to_numpy()
as_categorical_pclass = to_categorical(Pclass -1, num_classes=3)
as_categorical_pclass
```

```python
def sex_str_to_binary(s):
    if s=="male":
        return 1
    else:
        return 0
Sex=train_data['Sex']
# Sex.apply(sex_str_to_binary)
as_binary_sex=Sex.apply(sex_str_to_binary).to_numpy()
as_binary_sex
```

```python
train_Y=train_data['Survived']
```

```python
train_Y.to_numpy()
```

```python
age=train_data['Age'].to_numpy()
age=(age-train_data['Age'].mean()/train_data['Age'].std())
age[:10]
```

```python
sibsp=train_data['SibSp']
sibsp=sibsp.to_numpy()
sibsp=(sibsp-train_data['SibSp'].mean())/train_data['SibSp'].std()
sibsp[:10]
# sibsp=sibsp.value_counts()
```

```python
parch=train_data['Parch']
parch=parch.to_numpy()
parch=(parch-train_data['Parch'].mean())/train_data['Parch'].std()
parch[:10]
# parch=parch.value_counts()
```

```python
fare=train_data['Fare']
fare=fare.to_numpy()
fare=(fare-train_data['Fare'].mean())/train_data['Fare'].std()
fare[:10]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
train_X=np.concatenate((as_categorical_pclass,
                as_binary_sex.reshape(-1,1),
               age.reshape(-1,1),
               sibsp.reshape(-1,1),
                parch.reshape(-1,1),
                fare.reshape(-1,1)), axis=1)
train_X
```

```python
model=LogisticRegression(random_state=0).fit(train_X,train_Y)
```

```python
model.score(train_X,train_Y)
```
