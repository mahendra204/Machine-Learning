```python
featured_factors=['condition','grade','yr_built','floors','sqft_living']
x_input=house_data[featured_factors]
y_target=house_data['price']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train_x,value_x,train_y,value_y=train_test_split(x_input,y_target,test_size=0.4, random_state=42)
linear_model=LinearRegression()
linear_model.fit(train_x,train_y)
```

```python
from sklearn.metrics import mean_absolute_error
val_predictions=linear_model.predict(value_x)
print(mean_absolute_error(value_y,val_predictions))
```
