```python
data=pd.read_csv('/kaggle/input/spaceship-titanic/train.csv')
```

```python
features=['HomePlanet', 'CryoSleep', 'Destination', 'Age',
       'VIP', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
```

```python
train_data=data.drop('Transported', axis=1)
```

```python
train_data
```

```python
train_data=train_data[features]
```

```python
train_data=pd.get_dummies(train_data).fillna(0)
```

```python
train_data
```

```python
y_train=data['Transported']
```

```python
y_train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_val, y_train, y_val=train_test_split(train_data, y_train, train_size=0.9, test_size=0.1, random_state=1)
```

```python
x_train.shape
```

```python
x_val.shape
```

```python
y_train.shape
```

```python
y_val.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
d_model=DecisionTreeClassifier(random_state=42)
r_model=RandomForestClassifier(n_estimators=500, max_depth=9, random_state=1)
lr_model=LogisticRegression()
```

```python
d_model=d_model.fit(x_train,y_train)
r_model=r_model.fit(x_train,y_train)
lr_model=lr_model.fit(x_train,y_train)
```

```python
dpred=d_model.predict(x_val)
rpred=r_model.predict(x_val)
lrpred=lr_model.predict(x_val)
```

```python
dpred.shape
```

```python
rpred.shape
```

```python
lrpred.shape
```

```python
y_val.shape
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
metrics = {
    "Decision Tree": {
        "Accuracy": accuracy_score(y_val, dpred),
        "Precision": precision_score(y_val, dpred, average='weighted'),
        "Recall": recall_score(y_val,dpred, average='weighted'),
        "F1 Score": f1_score(y_val, dpred, average='weighted')
    },
    "Random Forest": {
        "Accuracy": accuracy_score(y_val,rpred),
        "Precision": precision_score(y_val,rpred, average='weighted'),
        "Recall": recall_score(y_val, rpred, average='weighted'),
        "F1 Score": f1_score(y_val, rpred, average='weighted')
    },
    "Logistic Regression": {
        "Accuracy": accuracy_score(y_val, lrpred),
        "Precision": precision_score(y_val,lrpred, average='weighted'),
        "Recall": recall_score(y_val, lrpred, average='weighted'),
        "F1 Score": f1_score(y_val, lrpred, average='weighted')
    }
}

# Print metrics
for classifier, metrics_dict in metrics.items():
    print(f"{classifier} Metrics:")
    for metric, value in metrics_dict.items():
        print(f"{metric}: {value:.4f}")
    print()
```
