```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("Assignment.csv")
```

```python
traindata=data.drop(['NO3','F'], axis=1)
```

```python
traindata
```

```python
traindata.isnull().sum()
```

```python
traindata=traindata.dropna()
```

```python
targetdata=data[['NO3','F']]
```

```python
targetdata=targetdata.dropna()
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(traindata, targetdata, test_size=0.2, random_state=1)
```

```python
print(xtrain.shape, ytrain.shape, xtest.shape, ytest.shape)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
model.fit(xtrain, ytrain)
```

```python
res=model.predict(xtest)
```

```python
res
```

```python
result = pd.DataFrame(res)
```

```python
result
```

```python
mse=mean_squared_error(ytest, res)
```
